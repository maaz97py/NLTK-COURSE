```python
#gutenberg corpus

from nltk.corpus import gutenberg
print(gutenberg.fileids())
```

```python
print("Total words:", len(emma_words))
print("Vocabulary size:",len(set(emma_words)) )
```

```python
from nltk.corpus import brown

print(brown.categories())
```

```python
news_words = brown.words(categories='news')
print(news_words[:50])
```

```python
#wordnet

from nltk.corpus import wordnet as wn

synsets = wn.synsets("bank")

print(synsets)
```

```python
bank = synsets[4]

print("Name:", bank.name())

print("Definition:", bank.definition())

print("Examples:", bank.examples())

```

```python
for lemma in bank.lemmas():
    print(lemma.name())
```

```python
dog = wn.synset('dog.n.01')
cat = wn.synset('cat.n.01')
car = wn.synset('car.n.01')
```

```python
print("dog vs cat:", dog.path_similarity(cat))
print("dog vs car:", dog.path_similarity(car))
```

```python
print(dog.wup_similarity(cat))
print(dog.lch_similarity(cat))
```
